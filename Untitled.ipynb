```python
%%sql
SELECT *
FROM competition
JOIN proposal ON competition.competitionID = proposal.competitionID
WHERE area = 'Humanities' 
```

```python
%%sql
SELECT AVG(ABS(proposal.requestedAmount - proposal.awardAmount))
FROM proposal
JOIN competition 
WHERE competition.area = 'Humanities'

```

```python
%%sql

SELECT AVG(ABS(proposal.requestedAmount-awardAmount))
FROM proposal
WHERE competitionID IN (SELECT competitionID FROM competition WHERE area COLLATE NOCASE = 'Humanities')
```
